**Pre-Processing Data**

Preprocessing data merupakan tahapan awal dalam proses analisa data yang bertujuan untuk membersihkan data, mengubah data  dan mempersiapkan data tersebut agar siap digunakan untuk analisa lebih lanjut.

```python
# Cek apakah terdapat nilai yg kosong pada tabel course
print("Course info")
course.isnull().sum()
```

```python
# Cek apakah terdapat nilai yg kosong pada tabel assesment info
print("Assesment info")
assesment.isnull().sum()
```

```python
#Handling Missing Value
#Mengisis data yang kosong pada kolom date di tabel Assessment dengan nilai rata-ratanya
assesment['date'].fillna(assesment['date'].mean(), inplace=True)
```

```python
# Cek apakah terdapat nilai yg kosong pada tabel vle info
print("Vle info")
vle.isnull().sum()
```

```python
#Handling Missing Value
#Handling missig value pada kolom week_from dan week_to di tabel Vle
#dengan menghapus kolomnya karena data yang kosong terlalu banyak
del vle['week_from']
del vle['week_to']
```

```python
# Cek apakah terdapat nilai yg kosong pada tabel student info
print("Student info")
info_student.isnull().sum()
```

```python
#Handling Missing Value
#Mengisi data yang kosong pada kolom imd_band di tabel Student Info dengan nilai modusnya
info_student['imd_band'].fillna(info_student['imd_band'].mode(), inplace=True)
```

```python
# Cek apakah terdapat nilai yg kosong pada tabel Student registration info
print("Reg Student info")
reg_student.isnull().sum()
```

```python
#Handling Missing Value
#Handling missig value pada kolom date_registration dan date_unregistration di tabel Student Registration
#dengan menghapus kolomnya karena data yang kosong terlalu banyak
del reg_student['date_registration']
del reg_student['date_unregistration']
```

```python
# Cek apakah terdapat nilai yg kosong pada tabel Asses student
print("Asses Student info")
asses_student.isnull().sum()
```

```python
#Handling Missing Value
#Menghapus baris yang kosong pada kolom score di tabel Student Assessment
asses_student.dropna(subset=['score'], inplace=True)
```

```python
# Cek apakah terdapat nilai yg kosong pada tabel Vle student
print("Vle Student info")
vle_student.isnull().sum()
```

**Merging Dataset**
Merging dataset merupakan proses menggabungkan dua atau lebih data berdasarkan satu atau beberapa variabel yang memiliki nama yang samadi setiap data. Adapun tujuan dari merge data ini adalah utuk mengganbungkan informasi dari data yang berbeda menjadi 1 data yang lebih lengkap.

Integrasi 2 variabel (student info dan assesment student)

```python
#Penggabungan dataset onfo_student dan asses_student, berdasarkan kolom id_student
data = pd.merge(info_student, asses_student, on = 'id_student')
```

```python
data.head()
```

```python
# Cek apakah terdapat nilai yang kosong dan mengetahui tipe data dari semua atribut
data.info()
```

```python
# Cek apakah terdapat nilai yg kosong
data.isnull().sum()
```

```python
##Menghapus nilai yang kosong
imd_band_miss = data['imd_band'].mode()[0]
score_miss = data['score'].mode()[0]
data['imd_band'].fillna(imd_band_miss, inplace=True)
data['score'].fillna(score_miss, inplace=True)
print(data.isnull().sum())
```

membuat hasil biner untuk analisis fitur

```python
data['final_result_2'] = data['final_result'].apply(lambda x: 'Failed' if \
    (x == 'Withdrawn' or x == 'Fail') else 'Succeeded')
```

##Gender Analysis

```python
#menampilkan isi dari jenis gender
data['gender'].unique()
```

```python
gender = data[['id_student', 'gender']].drop_duplicates().groupby('gender').\
    agg({'id_student':'count'}).reset_index().\
    rename(columns={"gender": "Gender", 'id_student': 'Quantity'})

gender['Percentual'] = gender.apply(lambda x: round(100 * (x['Quantity'] / gender['Quantity'].sum()),2), axis=1)

fig = px.bar(gender, x='Gender', y='Quantity', \
    text='Quantity', color='Gender',  hover_data=['Quantity'],\
        template="seaborn")

fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Distribusi - Gender',
    xaxis_title="Gender",
    yaxis_title="Percentual [%]",
    legend_title="Gender",
    width=600,
    height=400,
    uniformtext_minsize=10,
    uniformtext_mode='hide',)
fig.show()
```

- Menampilkan bahwa gender Male mempunyai jumlah yang lebih banyak dibandingkan dengan Female

```python
#Menampilkan total maisng masing gender berdasarkan quantity dan persentase
gender
```

Hubungan antara gender dan hasil akhir

```python
df = data.groupby(['gender', 'final_result']).agg({'id_student':'count'}).reset_index().\
    rename(columns={"gender": "Gender", "final_result": 'Result','id_student': 'Quantity'})

df['Percentual'] = df.apply(lambda x:  round(100 * (x['Quantity'] / df[df['Result'] == x['Result']]['Quantity'].sum()),2), axis=1)

fig = px.bar(df, x="Result", y="Percentual", text="Percentual", color="Gender",
            hover_data=['Percentual'], barmode = 'stack', template="seaborn")

fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Hasil akhir berdasarkan Gender',
    xaxis_title="Result",
    yaxis_title="Percentual [%]",
    legend_title="Gender",
    width=600,
    height=400,
    uniformtext_minsize=10,
    uniformtext_mode='hide',
)
fig.show()
```

- Pada diagram di atas diketahui bahwa untuk jenis kategori Distinction, laki-laki mempunyai jumlah yang lebih banyak dibandingkan perempuan.

- Pada kategori Fail, laki-laki mempunyai jumlah siswa yang lebih banyak dibandingkan dengan perempuan

- Pada kategori Pass juga laki-laki mempunyai jumlah siswa yang lebih banyak dibandingkan dengan perempuan

- Pada kategori Withdrawn juga laki-laki mempunyai jumlah siswa yang lebih banyak dibandingkan dengan perempuan

yang mana kita simpulkan bahwa laki laki mempunyai performa hasil akhir yang lebih bagus dibandingkan dengan perempuan

```python
# Hasil Biner - Lulus dan Gagal
# ----------------------------------------------------------------------------------------------------
# ----------------------------------------------------------------------------------------------------

df = data.groupby(['gender', 'final_result_2']).agg({'id_student':'count'}).reset_index().\
    rename(columns={"gender": "Gender", "final_result_2": 'Result','id_student': 'Quantity'})

df['Percentual'] = df.apply(lambda x:  round(100 * (x['Quantity'] / df[df['Result'] == x['Result']]['Quantity'].sum()),2), axis=1)

fig = px.bar(df, x="Result", y="Percentual", text="Percentual", color="Gender",
            hover_data=['Percentual'], barmode = 'stack', template="seaborn")

fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Hasil berdasarkan Gender',
    xaxis_title="Result",
    yaxis_title="Percentual [%]",
    legend_title="Gender",
    width=600,
    height=400,
    uniformtext_minsize=10,
    uniformtext_mode='hide',
)
fig.show()
```

##Region Analysis

```python
#Menampilkan jumlah kategori region
len(data['region'].unique())
```

```python
region = data[['region', 'id_student']].drop_duplicates().groupby('region').agg({'id_student':'count'}).reset_index().\
    rename(columns={"region": "Region", 'id_student': 'Quantity'}).sort_values(by='Quantity', ascending=False)

region['Percentual'] = region.apply(lambda x: round(100 * (x['Quantity'] / region['Quantity'].sum()), 2) ,axis=1)

fig = px.bar(region, x='Region', y='Quantity', text='Percentual', color ='Region',  template="seaborn")
fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Distribusi - Lokasi',
    xaxis_title="Region",
    yaxis_title="Quantity [siswa]",
    showlegend=False,
    width=800,
    height=550,
    uniformtext_minsize=10,
    uniformtext_mode='hide',
)

fig.show()
```

- dari data di atas, diketahui bahwa jumlah frekuensi student terbanyak berasal dari region Scotland dan jumlah frekuensi student terdikit ada yang berasal dari region Ireland.

```python
df = data.groupby(['region', 'final_result']).agg({'id_student':'count'}).reset_index().\
    rename(columns={"region": "Region", "final_result": 'Result','id_student': 'Quantity'})

df['Percentual'] = df.apply(lambda x:  round(100 * (x['Quantity'] / df[df['Result'] == x['Result']]['Quantity'].sum()),2), axis=1)

fig = px.bar(df, x="Result", y="Percentual", text="Percentual", color="Region",
            hover_data=['Percentual'], barmode = 'stack', template='seaborn')
fig.update_layout(
    margin=dict(l=20, r=50, t=50, b=20),
    title = 'Hasil berdasarkan Lokasi',
    xaxis_title="Result",
    yaxis_title="Percentual [%]",
    legend_title="Region",
    width=800,
    height=700,
    uniformtext_minsize=9,
    uniformtext_mode='hide',
)

fig.show()
```
